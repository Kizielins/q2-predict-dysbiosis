```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the tab-delimited file
data = pd.read_csv("boruta_results/boruta_importance.txt", sep="\t", index_col=0)

# Ensure the data is numeric
data = data.apply(pd.to_numeric, errors='coerce')

# Sort the index alphabetically
data = data.sort_index()

# Create a heatmap
plt.figure() #figsize=(12, 8))  # Adjust figure size
sns.heatmap(
    data,
    cmap="RdYlGn_r",  # Reverse Red-Yellow-Green for color mapping
    annot=True,       # Show rank values in the heatmap
    cbar_kws={'label': 'Importance Rank'},  # Add colorbar label
    fmt=".0f"         # Format annotations as integers
)

# Add title and labels
plt.title("Feature Ranks per Cohort", pad=20)
plt.xlabel("")
plt.ylabel("")


# Show the plot
plt.show()
```
